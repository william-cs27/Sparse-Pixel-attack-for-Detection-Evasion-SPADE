```python
import os
from ultralytics import YOLO

# ==== USER INPUT ====
model_path = '/content/drive/MyDrive/runs/detect/train5/weights/best.pt'          # Path to your trained YOLOv8n model
input_dir = '/content/drive/MyDrive/Project_RLAB/RFPAR/Results/adv_images'           # Directory with images to predict
output_dir = '/content/drive/MyDrive/Project_RLAB/RFPAR/Results/pred_results'   # New directory for results

# ==== MAKE OUTPUT DIR IF IT DOESN'T EXIST ====
os.makedirs(output_dir, exist_ok=True)

# ==== LOAD MODEL ====
model = YOLO(model_path)

# ==== GET LIST OF IMAGES ====
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff']
image_files = [os.path.join(input_dir, f)
               for f in os.listdir(input_dir)
               if os.path.splitext(f)[1].lower() in image_extensions]

# ==== RUN PREDICTION ====
results = model.predict(
    source=image_files,
    save=True,                      # Save output images with predictions
    save_txt=True,                  # Also save .txt files with results
    project=output_dir,             # Base output folder
    name='',                        # Saves directly into output_dir
    exist_ok=True                   # Overwrite existing results if needed
)

print(f"Predictions saved to: {output_dir}")

```

```python
import os
from ultralytics import YOLO

# ==== USER INPUT ====
model_path = '/content/drive/MyDrive/runs/detect/train5/weights/best.pt'          # Path to your trained YOLOv8n model
input_dir = '/content/drive/MyDrive/Project_RLAB/Dataset/BD100K/images/test/'   # Directory with images to predict
output_dir = '/content/drive/MyDrive/Project_RLAB/RFPAR/Results/org_pred_results' # New directory for results

# ==== MAKE OUTPUT DIR IF IT DOESN'T EXIST ====
os.makedirs(output_dir, exist_ok=True)

# ==== LOAD MODEL ====
model = YOLO(model_path)

image_files = [f for f in os.listdir(input_dir)
               if os.path.splitext(f)[1].lower() in image_extensions]

image_files = sorted(image_files)    # <-- Sort the filenames

# ==== SELECT ONLY FIRST 100 IMAGES ====
image_files = image_files[:100]

# Get full paths
image_files = [os.path.join(input_dir, f) for f in image_files]


# ==== RUN PREDICTION ====
results = model.predict(
    source=image_files,
    save=True,                      # Save output images with predictions
    save_txt=True,                  # Also save .txt files with results
    project=output_dir,             # Base output folder
    name='',                        # Saves directly into output_dir
    exist_ok=True                   # Overwrite existing results if needed
)

print(f"Predictions saved to: {output_dir}")
```
